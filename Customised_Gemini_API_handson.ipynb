```python
import os
import google.generativeai as genai

# Use REST (avoids slow gRPC issues)
os.environ["GOOGLE_API_USE_REST"] = "true"

genai.configure(api_key="")

model = genai.GenerativeModel("models/gemini-2.5-flash")

# System behavior (manual)
system_msg = input("What type of chatbot would you like to create?\n")

chat = model.start_chat(history=[])

print("Your assistant is ready! Type 'quit' to exit.\n")

while True:
    user_input = input("You: ")

    if user_input.lower() == "quit":
        break

    # Combine system instruction + user input
    prompt = f"{system_msg}\nUser: {user_input}"

    response = chat.send_message(prompt)

    print("Bot:", response.text, "\n")
```
