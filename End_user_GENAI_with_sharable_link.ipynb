```python
import google.generativeai as genai
import gradio as gr

genai.configure(api_key="")

model = genai.GenerativeModel("gemini-pro")

# conversation memory
messages = [{"role": "system", "parts": ["You are a financial expert specializing in real estate investment and negotiation"]}]

def CustomGemini(user_input):
    # add user message
    messages.append({"role": "user", "parts": [user_input]})
    
    # generate response
    response = model.generate_content(messages)
    
    reply = response.text
    
    # store model reply
    messages.append({"role": "model", "parts": [reply]})
    
    return reply

demo = gr.Interface(
    fn=CustomGemini,
    inputs="text",
    outputs="text",
    title="My Saving Assistant"
)

demo.launch(share=True)
```
