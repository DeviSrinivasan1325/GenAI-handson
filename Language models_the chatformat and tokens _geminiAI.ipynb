```python
import google.generativeai as genai

genai.configure(api_key="")

def get_completion(prompt,model='models/gemini-2.5-flash'):
    model = genai.GenerativeModel(model)
    messages = [{"role":"user","parts":[prompt]}]
    response = model.generate_content(messages,generation_config={"temperature": 0})
    return response.text
```

```python
response = get_completion("Take the letter in lollipop and reverse them")
print(response)
```

## Chat format (helper function)

```python
def get_completion_from_message(messages,model='models/gemini-2.5-flash',temperature=0,max_output_tokens = 500):
    model = genai.GenerativeModel(model)
    response = model.generate_content(messages,generation_config={"temperature": 0,"max_output_tokens": max_output_tokens})
    return response.text
messages=[{"role":"user",
           "parts":["you are an assistant who responds in the style of Dr.Seuss."]},
         {"role":"user",
          "parts":["write me a very short poem about a happy carrot"]}]
```

```python
response = get_completion_from_message(messages,temperature = 1)
print(response)
```
